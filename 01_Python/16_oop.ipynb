```python
class Employee:
    company="HP"
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
        
    def printDetails(self):
        print(f"The company of {self.name} is {self.company} with salary ${self.salary}")

    @staticmethod     # tells a method that it does not need self or cls. it is a standalone method. it can be called independently
    def printTime():
        print(f"The time is now")

    @classmethod     #can access class related attributes only & not object related
    def print_class_details(cls):
        print(f"The company is {cls.company}")

e=Employee("John",47000)
e.printDetails()
e.printTime()
e.print_class_details()
Employee.print_class_details()
```

```python
import json   
```

```python
obj={                  # python dictionary
    "name":"Alice",
    "company":"IBM",
    "salary":45000
}
json.dumps(obj) # dumps - converts a python object to json string
```

```python
d=json.loads('{"name": "Alice", "company": "IBM", "salary": 45000}')  # loads - converts a json string to python object
```

```python
type(d)
```

# json.dump()  -  Writes JSON data to a file

```python
file=open("data.json","w")
```

```python
json.dump(obj,file) # (data,file)
```

```python
file.close()
```

# Formatting JSON Output

```python
formatted_json=json.dumps(obj,indent=4)
```

```python
print(formatted_json)
```

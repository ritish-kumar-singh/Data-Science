```python
import json
```

```python
def load_data(filename):
    with open(filename,'r') as f:
        return json.load(f)
```

```python
def find_people_you_may_know(user_id, data):
    user_friends = {}
    for user in data['users']:
        user_friends[user['id']] = set(user['friends'])
    
    if user_id not in user_friends: # if no friends exist
        return []

    direct_friends = user_friends[user_id]
    suggestions = {}
    for friend in direct_friends:
        for mutual in user_friends[friend]:
            if(mutual!=user_id and mutual not in direct_friends):
                # count mutual friends
                suggestions[mutual] = suggestions.get(mutual,0) + 1

    sorted_suggestions = sorted(suggestions.items(), key=lambda x: x[1], reverse=True) # sorted in descending on the basis of element at idx 1, i.e. value
    return [user_id for user_id, mutual_count in sorted_suggestions]
            
```

```python
# load the data
data=load_data('massive_data.json')
user_id=1
recommendations=find_people_you_may_know(user_id,data)
print(recommendations)
```
